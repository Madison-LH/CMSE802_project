```python

import matplotlib.pyplot as plt

# Data
labels = ['INSC', 'PERC', 'NOBR', 'ENTR', 'EXIT']
counts = [875793, 42620, 12788, 89, 271]

# Increase explode values especially for the smaller slices
explode = (0, 0.05, 0.1, 0.3, 0.4)  # bigger explode for ENTR & EXIT

plt.figure(figsize=(10, 10))
plt.pie(
    counts,
    labels=labels,
    autopct='%1.1f%%',
    startangle=140,
    labeldistance=1.3,
    pctdistance=0.8,
    explode=explode
)
plt.title("Total Frame Counts Across All Videos - Pie Chart")
plt.axis('equal')
plt.tight_layout()
plt.show()

# Bar Chart
plt.figure(figsize=(10, 6))
bars = plt.bar(labels, counts, color='skyblue')
plt.title("Total Frame Counts Across All Videos - Bar Chart")
plt.xlabel("Frame Category")
plt.ylabel("Frame Count")

# Add the data labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, height, f'{height}', 
             ha='center', va='bottom')

plt.show()

```
